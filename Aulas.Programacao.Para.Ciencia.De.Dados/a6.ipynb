```python
import matplotlib.pyplot as plt
import numpy as np

# Cenário em que só queremos mostrar várias figura
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sin(2*x)
y5 = np.cos(2*x)
y6 = np.tan(2*x)
# Definimos a figura
fig, axs = plt.subplots(2, 3, figsize=(10,7.5))

axs[0,0].plot(x,y1,label='sen(x)')
axs[0,0].set_title('Desenho do grafico do sen(x)')
axs[0,0].legend()

axs[0,1].plot(x,y2,label='cos(x)')
axs[0,1].set_title('Desenho do grafico do cos(x)')
axs[0,1].legend()

axs[0,2].plot(x,y3,label='tan(x)')
axs[0,2].set_title('Desenho do grafico do tan(x)')
axs[0,2].legend()

axs[1,0].plot(x,y4,label='sen(2*x)')
axs[1,0].set_title('Desenho do grafico do sen(2*x)')
axs[1,0].legend()

axs[1,1].plot(x,y5,label='cos(2*x)')
axs[1,1].set_title('Desenho do grafico do cos(2*x)')
axs[1,1].legend()

axs[1,2].plot(x,y6,label='tan(2*x)')
axs[1,2].set_title('Desenho do grafico do tan(2*x)')
axs[1,2].legend()

# Criamos a figura
plt.show()
```

```python
#Também podem definir-se os axes utilizando a instrução sub_plot
import matplotlib.pyplot as plt
import numpy as np

#Exemplo com duas sub figuras
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(1,2,1) #As posicoes são definidas da seguinte forma: 1 linha, 2 colunas, 1ª figura
ax2 = fig.add_subplot(1,2,2) #As posicoes são definidas da seguinte forma: 1 linha, 2 colunas, 2ª figura

ax1.plot(x,y1,label='sen(x)')
ax2.plot(x,y2,label='cos(x)')
ax1.legend()
ax2.legend()
plt.show()
```

```python
#1. Carregar os dados para um Pandas DataFrame e fazer algumas análises
import pandas as pd
import numpy as np
df1 = pd.read_csv('https://raw.githubusercontent.com/atrigo/PyTrigo/master/Versao2/titanic.csv')
#Ver as cinco primeiras linhas
display(df1.head())
##GUARDAR EM CSV NA PASTA:
df1.to_csv('tatinic.csv')
```

```python
##VERIFICAR MISSING VALUES 
df1.isnull().sum()
```

```python
#2. Reparem que temos valores com pontos de interrogação em algumas colunas.
#Estes valores, neste caso, são valores em falta.
#Vamos substituir os valores em falta por NaN (Not a Number)
df1.replace('?', np.nan, inplace=True)
display(df1.head())
```

```python
#3. Vamos verificar os tipos de dados, pois como tínhamos valores em falta, o Pandas pode ter assumido tipos de dados errados.
display(df1.dtypes)
```

```python
#4. Reparem que as colunas age e fare são do tipo object quando deveriam ser do tipo float64, vamos convertê-las no tipo float64 e ver o que aontece
df1['age'] = pd.to_numeric(df1['age'], errors='coerce')
df1['fare'] = pd.to_numeric(df1['fare'], errors='coerce')
display(df1.dtypes)
```

```python
#5. Vamos verificar se existem valores em falta, agora que já não temos o ? e temos o NaN (Not a Number)
display(df1.isnull().sum())
```

```python
#6. O tratamento dos valores em falta depende do que queremos analisar
#Se não quisermos analisar a cabin, boat e o body, podemos eliminar as colunas
df1.drop(['cabin', 'boat', 'body'], axis=1, inplace=True)
display(df1.head())
```

```python
#7. Quanto ao valor da fare em falta, podemos substituir pela média da resptiva classe (pensando que a classe influencia o valor da fare, que podemos verificar mais à frente)
class_fare_means = df1.groupby('pclass')['fare'].mean()
df1['fare'] = df1.apply(lambda row: class_fare_means[row['pclass']] if pd.isnull(row['fare']) else row['fare'], axis=1)
print(df1['fare'].isnull().sum())
```

```python
#8. Quanto às idades depende do que queremos analisar podemos substituir pela média das idades, pela mediana ou pela moda ou podemos eliminar as linhas com valores em falta
#Aqui vamos optar por elimnar as linhas com valores em falta, pois a ideia é fazer alguns exemplos de visualização de dados
df1.dropna(subset=['age'],inplace=True)
display(df1.isnull().sum())
#Caso quisessem remover todas as colunas onde existem valores nulos poderiam fazer:df1.dropna(inplace=True)
```

```python
#9. Podemos ainda remover os nulos do embarked e remover a coluna home.dest
df1.dropna(subset=['embarked'],inplace=True)
df1.drop(['home.dest'], axis=1, inplace=True)
display(df1.isnull().sum())
display(df1.head())
```

Existem várias formas de tratar os outliers e mesmo utilizando essas diferentes formas é preciso ter cuidado, pois podem eixistir outliers que não o são.

Uma forma commumente utilizada é a eliminação dos valores que estão a mais de 3 std da média (Three standard deviations from the mean is a common cut-off in practice for identifying outliers in a Gaussian or Gaussian-like distribution. For smaller samples of data, perhaps a value of 2 standard deviations (95%) can be used, and for larger samples, perhaps a value of 4 standard deviations (99.9%) can be used).

A eleminaçã ode outliers depende de cada caso em concreto e até pode haver necessidade de não elminar outliers

```python
#No exemplo anteriore vamos elminar os registos em que a tarifa é superior a 3 desvios padrão
#Vamos calcular os limites superior e inferior
print(df1.shape)
mean = df1['tarifa'].mean()
std = df1['tarifa'].std()
upper_limit = mean + 3*std
lower_limit = mean - 3*std
print('Upper limit:', upper_limit)
print('Lower limit:', lower_limit)
#Vamos eliminar os registos
df1 = df1[(df1['tarifa'] < upper_limit) & (df1['tarifa'] > lower_limit)]
#Vamos verificar os registos
print(df1.shape)

#Podemos voltar a ver o boxplot, sendo que a principal diferença é a de que o eixo x está mais reduzido
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.boxplot(x='tarifa', data=df1, color='skyblue')
plt.xlabel('Tarifa')
plt.title('Boxplot da tarifa')
plt.show()
```

## Criação das tabelas fictícias em ficheiros CSV

```python
# Para quem quiser criar as tabelas de origem poderá utilizar o seguinte código
#%pip install faker
#%pip install faker-commerce

#Criação da tabela clientes.csv
import pandas as pd
from faker import Faker
Faker.seed(10)
fake = Faker("pt_PT")
# Defina o número de clientes que deseja gerar
num_clientes = 100  # Você pode ajustar esse número conforme necessário
# Crie uma lista de dicionários com IDs e nomes gerados automaticamente
dados_clientes = []
for i in range(num_clientes):
    cliente = {
        'c_id': i + 1,  # IDs começando em 1
        'c_nome': fake.name()
    }
    dados_clientes.append(cliente)
# Crie um DataFrame do Pandas com os dados
df = pd.DataFrame(dados_clientes)
df.to_csv('clientes.csv', index=False)

#Criação da tabela produtos.csv
#import pandas as pd
#from faker import Faker
import faker_commerce

# Crie uma instância do Faker
Faker.seed(10)
fake = Faker("pt_PT")
fake.add_provider(faker_commerce.Provider)
# Defina o número de produtos que deseja gerar
num_produtos = 100  # Você pode ajustar esse número conforme necessário
# Crie uma lista de dicionários com IDs e nomes gerados automaticamente
dados_produtos = []
for i in range(num_produtos):
    produto = {
        'p_id': i + 1,  # IDs começando em 1
        'p_nome': fake.ecommerce_name(),
        'p_preco': fake.pyfloat(left_digits=2, right_digits=2, positive=True)
    }
    dados_produtos.append(produto)
# Crie um DataFrame do Pandas com os dados
df = pd.DataFrame(dados_produtos)
# Imprima o DataFrame
#print(df)
df.to_csv('produtos.csv', index=False)

#Criação da tabelas vendas
#import pandas as pd
#from faker import Faker
#import faker_commerce
# Crie uma instância do Faker
Faker.seed(10)
fake = Faker("pt_PT")
fake.add_provider(faker_commerce.Provider)

# Defina o número de clientes que deseja gerar
num_vendas = 1000  # Você pode ajustar esse número conforme necessário
# Crie uma lista de dicionários com IDs e nomes gerados automaticamente
dados_vendas = []
for i in range(num_vendas):
    venda = {
        's_id': i + 1,  # IDs começando em 1
        'c_id': fake.pyint(min_value=1, max_value=100),
        'p_id': fake.pyint(min_value=1, max_value=100),
        'qty' : fake.pyint(min_value=1, max_value=20),
        'store': fake.company()
    }
    dados_vendas.append(venda)
# Crie um DataFrame do Pandas com os dados
df = pd.DataFrame(dados_vendas)
# Imprima o DataFrame
#print(df)
df.to_csv('vendas.csv',index=False)
```

## Importação das tabelas CSV para o SQLite

```python
#Carregar os ficheiros csv para um DataFrame e enviar para a base de dados
#Carregar os ficheiros csv DataFrame
df_vendas = pd.read_csv('vendas.csv')
df_produtos = pd.read_csv('produtos.csv')
df_clientes = pd.read_csv('clientes.csv')
#Cria uma base de dados em sqlite e guardar as tabelas produtos, clientes e vendas
import sqlite3
conn = sqlite3.connect('vendas.sqlite')
df_clientes.to_sql('clientes', conn, if_exists='replace', index=False)
df_produtos.to_sql('produtos', conn, if_exists='replace', index=False)
df_vendas.to_sql('vendas', conn, if_exists='replace', index=False)
conn.close()
```

```python
#Aceder a dados existentes nas tabelas das bases de dados e realizar querys
#Atenção que pode carregar sempre os dados todos para um DataFrame e depois fazer querys
import sqlite3
conn = sqlite3.connect('vendas.sqlite')
#Selecionar todos os clientes para um DataFrame
df1 = pd.read_sql('select * from clientes', conn)
df1.head(10)
#Vendas
df2 = pd.read_sql('select * from vendas', conn)
display(df2.head(1)   )
#Vendas com informcao do produto
df3 = pd.read_sql('select * from vendas v inner join produtos p on v.p_id=p.p_id', conn) 
display(df3.head(1))
#Vendas por produto sem infromacao do produto
df4 = pd.read_sql('select p_id, sum(qty) from vendas group by p_id', conn)
display(df4.head(1))
#Vendas por produto com informacao do produto
df5 = pd.read_sql('select p.p_id, p.p_nome, sum(v.qty) from vendas v inner join produtos p on v.p_id=p.p_id group by p.p_id', conn)
display(df5.head(1))
#Vendas por produto com informacao do produto e total da venda
df6 = pd.read_sql('select p.p_id, p.p_nome, sum(v.qty), sum(v.qty)*p.p_preco from vendas v inner join produtos p on v.p_id=p.p_id group by p.p_id', conn)
display(df6.head(1))
#Vendas por produto por loja
df7 = pd.read_sql('select p.p_id, v.store, sum(v.qty) from vendas v inner join produtos p on v.p_id=p.p_id group by p.p_id, v.store', conn)
display(df7.head(1))
#Vendas por produto por loja com nome do produto
df8 = pd.read_sql('select p.p_id, p.p_nome, v.store, sum(v.qty) from vendas v inner join produtos p on v.p_id=p.p_id group by p.p_id, v.store', conn)
display(df8.head(1))
#Produtos que não foram vendidos
df9 = pd.read_sql('select p.p_id, p.p_nome from produtos p left join vendas v on p.p_id=v.p_id where v.p_id is null', conn)
display(df9.head(1))
#Clientes que não compraram
df10 = pd.read_sql('select c.c_id, c.c_nome from clientes c left join vendas v on c.c_id=v.c_id where v.c_id is null', conn)
display(df10.head(1))
#Fechar a ligacao
conn.close()
```
