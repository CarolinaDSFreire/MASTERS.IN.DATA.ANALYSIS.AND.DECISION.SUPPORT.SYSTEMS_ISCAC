```python
#1.3.3. #CHATGPT
def calculadora():
    while True:
        print("Operações disponíveis:")
        print("1. Adição")
        print("2. Subtração")
        print("3. Multiplicação")
        print("4. Divisão")
        print("5. Sair")
        
        escolha = input("Escolha a operação (1/2/3/4/5): ")

        if escolha == '5':
            print("Calculadora encerrada.")
            break
        elif escolha not in ('1', '2', '3', '4'):
            print("Escolha inválida. Tente novamente.")
            continue

        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))

        if escolha == '1':
            resultado = num1 + num2
            print(f"Resultado: {num1} + {num2} = {resultado}")
        elif escolha == '2':
            resultado = num1 - num2
            print(f"Resultado: {num1} - {num2} = {resultado}")
        elif escolha == '3':
            resultado = num1 * num2
            print(f"Resultado: {num1} * {num2} = {resultado}")
        elif escolha == '4':
            if num2 == 0:
                print("Divisão por zero não é permitida.")
            else:
                resultado = num1 / num2
                print(f"Resultado: {num1} / {num2} = {resultado}")

calculadora()
```

```python
#1.3.4. #CHATGPT
numero = int(input('qual numero?'))
def primo(numero):
    if numero <= 1:
        return False  # Números menores ou iguais a 1 não são primos
    for i in range(2, numero):
        if numero % i == 0:
            return False  # Se for divisível por algum número, não é primo
    return True  # Se não for divisível por nenhum número, é primo

for ele in range(11, 91):
    if primo(ele):
        print(f'O número {ele} é primo')
```

```python
#1.3.5. #CHATGPT
import random

def dado():
    resultados = [0, 0, 0, 0, 0, 0]  # Uma lista para contar quantas vezes cada número aparece

    for i in range(100):
        resultado_lancamento = random.randint(1, 6)  # Simula o lançamento de um dado de 6 lados
        resultados[resultado_lancamento - 1] += 1  # Conta o resultado do lançamento

    total_lancamentos = sum(resultados)  # Total de lançamentos

    for i in range(6):
        percentual = (resultados[i] / total_lancamentos) * 100
        print(f"Número {i + 1} saiu {resultados[i]} vezes, o que representa {percentual:.2f}% dos lançamentos")

dado()
```

```python
#1.3.6. #FEITO EM SALA
numero = int(input('qual numero?'))
def fatorial(numero):
    if numero == 1 or numero == 0:
        return 1
    fatorial = 1
    for i in range(1,numero):
        fatorial*=i
    return fatorial

for ele in range(numero,16):
    print(f'{ele}!= {fatorial(ele)}')
```
