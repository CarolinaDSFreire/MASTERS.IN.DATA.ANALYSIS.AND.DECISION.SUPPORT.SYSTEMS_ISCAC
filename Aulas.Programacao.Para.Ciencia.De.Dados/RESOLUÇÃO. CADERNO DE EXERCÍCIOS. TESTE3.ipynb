```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df_titanic = sns.load_dataset('titanic')
```

```python
#4.4.1.
plt.hist(df_titanic['age'])
```

```python
#4.4.2. #TIAGO
pclass_data = np.random.choice(df_titanic['pclass'], 5)

bin_edges =[0.5, 1.5, 2.5, 3.5, 4.5]

custom_labels = [1, 2, 3]

plt.hist(pclass_data, bins=bin_edges, edgecolor='k')

plt.xlabel('Classe')
plt.ylabel('Frequência')

plt.xticks(custom_labels)

plt.title('Histograma de Classes')
```

```python
#4.4.3. #CHATGPT

import matplotlib.pyplot as plt
import numpy as np

# Gere alguns dados aleatórios para demonstração
data = np.random.randn(1000)

# Calcule as contagens e as posições dos bins
counts, bin_edges = np.histogram(data, bins=30)
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

# Plote as barras centralizadas nos valores das abscissas
plt.bar(bin_centers, counts, width=bin_edges[1] - bin_edges[0], edgecolor='black', alpha=0.7)

# Adicione rótulos e título
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma com Barras Centralizadas nos Valores das Abscissas')

# Mostre o gráfico
plt.show()

#bin_centers são os pontos médios entre os limites dos bins. 
#A função plt.bar é usada para plotar barras centradas nesses pontos médios. 
#O parâmetro width é definido como a largura dos bins, garantindo que as barras estejam centradas nos valores das abscissas.
```

```python
#F)

fig,ax = plt.subplots(nrows=2,ncols=2,dpi=120,figsize = (8,6), gridspec_kw={'hspace': 0.4, 'wspace': 0.4})

sns.distplot(df_titanic['age'], ax=ax[0][0])
ax[0][0].set_title('DistPlot da Idade', fontdict={'fontsize':8})
ax[0][0].set_xlabel('Idade',fontdict={'fontsize':7})

sns.boxplot(df_titanic['age'], ax=ax[0][1])
ax[0][1].set_title('BoxPlot da Idade', fontdict={'fontsize':8})
ax[0][1].set_xlabel('Idade',fontdict={'fontsize':7})

sns.distplot(df_titanic['fare'], ax=ax[1][0], color = 'green')
ax[1][0].set_title('DistPlot da Tarifa', fontdict={'fontsize':8})
ax[1][0].set_xlabel('Tarifa',fontdict={'fontsize':7})

sns.boxplot(df_titanic['fare'], ax=ax[1][1], color = 'green')
ax[1][1].set_title('BoxPlot da Tarifa', fontdict={'fontsize':8})
ax[1][1].set_xlabel('Tarifa',fontdict={'fontsize':7})
```

```python
#OU #THIAGO
# Crie uma figura e uma única linha com dois subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Gráfico de distribuição (distplot) da variável 'age'
sns.distplot(df_titanic['age'].dropna(), bins=20, ax=ax1)
ax1.set_title('Distribuição de Idades')
ax1.set_xlabel('Idade')

# Gráfico de caixa (boxplot) da variável 'fare'
sns.boxplot(y='fare', data=df_titanic, ax=ax2)
ax2.set_title('Boxplot de Fare')
ax2.set_ylabel('Fare')

# Ajuste de layout para evitar sobreposição de rótulos
plt.tight_layout()

# Mostrar os gráficos
plt.show()
```

```python
#G)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(12,6))

sns.countplot(x='survived', data=df_titanic, ax=ax1, color='green')
ax1.set_title('Número de sobreviventes')
ax1.set_xlabel('Sobreviventes')

sns.countplot(x='sex', data=df_titanic, ax=ax2, color='pink')
ax2.set_title('Passageiros por gênero')
ax2.set_xlabel('Gênero')

sns.countplot(x='pclass', data=df_titanic, ax=ax3)
ax3.set_title('Passageiros por classe')
ax3.set_xlabel('Classe')

sns.countplot(x='sex', hue= 'pclass', data=df_titanic, ax=ax4)
ax4.set_title('Passageiros por classe e gênero')
ax4.set_xlabel('Gêneros')
```

```python
#H)
sns.boxplot(x='pclass', y='fare', data=df_titanic)
plt.title('Boxplot da Tarifa por Pclass')
plt.xlabel('Pclass (Classe de Passageiro)')
plt.ylabel('Tarifa')
plt.show()
```

```python
#M)
import pandas as pd
faixas_etarias = [0, 10, 20, 30, 40, 50, 60, 70, 80]
rotulos = np.array(['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-80'])
df_titanic['faixa_etaria'] = pd.cut(df_titanic['age'], bins=faixas_etarias, labels=rotulos)

crosstab = pd.crosstab(df_titanic['faixa_etaria'], df_titanic['survived'], normalize='index')
crosstab.rename(columns={0: 'Não Sobreviveu', 1: 'Sobreviveu'}, inplace=True)
print(crosstab)
```

```python
sns.barplot(x=crosstab.index, y='Sobreviveu', data = crosstab)
plt.xlabel('Faixa Etária')
plt.ylabel('Taxa de Sobrevivência')
plt.title('Taxas de Sobrevivência por Grupo de Idade')
```
